```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
### Independent features and dependent feature
X = pd.DataFrame(iris['data'], columns=['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm'])
y = iris['target']
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
### Apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
treeclassifier.fit(X_train, y_train)
```

```python
### Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
acc = accuracy_score(y_test, y_pred)
print(acc)
cr = classification_report(y_test, y_pred)
print(cr)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.decomposition import PCA

# Giảm số đặc trưng xuống còn 2
pca = PCA(n_components=2)
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

# Decision Tree Classifier Prepruning And Hyperparameter Turning

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [i for i in range (1, 10)],
    'max_features': ['auto', 'sqrt', 'log2'],
}
```

```python
params
```

```python
treemodel = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=treemodel , param_grid=params, cv=5, verbose=3, refit=True, scoring='accuracy')
grid.fit(X_train_reduced, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best = grid.best_estimator_
```

```python
# Visualising the Test set results
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

X_set, y_set = X_test_reduced, y_test
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
    np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01)
)

plt.contourf(X1, X2, best.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(['#800080', '#FFA500']))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(['#FF0000', '#00FF00', '#0000FF'])(i), label=j)

plt.title('Decision Tree with PCA (Test set)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
