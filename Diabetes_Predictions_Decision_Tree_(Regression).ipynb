```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(dataset.data, columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
```

```python
df_diabetes.head()
```

```python
### Independent and Dependent features
X = df_diabetes
y = dataset['target']
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
X_train.head()
```

```python
### Correlation
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [i % 5 == 0 for i in range(5, 100)],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=regressor, param_grid=param, cv=5, verbose=3, n_jobs=-1,
                    refit=True, scoring='neg_mean_squared_error')
grid.fit(X_train, y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
score = r2_score(y_test, y_pred)
print(score)
mae = mean_absolute_error(y_test, y_pred)
print(mae)
mse = mean_squared_error(y_test, y_pred)
print(mse)
rmse = np.sqrt(mse)
print(rmse)
```

```python
selected_model = DecisionTreeRegressor(criterion='friedman_mse', max_depth=4, max_features='log2', splitter='best')
selected_model.fit(X_train, y_train)
```

```python
# Visualize the Decision Tree

import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(selected_model, filled=True)
```
